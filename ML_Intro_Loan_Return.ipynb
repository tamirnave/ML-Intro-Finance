```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

## Load the Database

```python
train_df = pd.read_csv('2019loans.csv')
test_df = pd.read_csv('2020Q1loans.csv')
```

```python
# Display the dataframe
train_df.head()
```

```python
test_df.head()
```

## Extract The Labels

```python
y_train = train_df["loan_status"]                       # Take target feature for training data
X_train = train_df.drop(columns = ["loan_status"])      # Drop target from training data

# Same for testing:
y_test = test_df["loan_status"]
X_test = test_df.drop(columns = ["loan_status"])
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Convert categorical variable into indicator variables

```python
X_train = pd.get_dummies(X_train) 
X_test = pd.get_dummies(X_test)
```

```python
# The get_dummies converted to a new binary column
X_train[['home_ownership_RENT', 'home_ownership_MORTGAGE', 'home_ownership_OWN']]
```

```python
# add missing dummy variables to testing set (in training only)
for column in X_train.columns:
    if column not in X_test.columns:
        X_test[column] = 0
        print(column)
```

```python
# Class distribution
print(sum(y_train=="low_risk") / len(y_train))
print(sum(y_test=="low_risk") / len(y_test))
```

## Scale the data

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# KNN

```python
model = KNeighborsClassifier(n_neighbors=12)

model.fit(X_train_scaled, y_train)

train_accuracy = model.score(X_train_scaled, y_train)
test_accuracy = model.score(X_test_scaled, y_test)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```

## Logistic Regression

```python
model = LogisticRegression(max_iter=1000)

model.fit(X_train_scaled, y_train)

train_accuracy = model.score(X_train_scaled, y_train)
test_accuracy = model.score(X_test_scaled, y_test)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```
